```python
from collections import defaultdict
N, M = map(int, input().split())
edges = defaultdict(list)
for _ in range(M):
    A, B = map(int, input().split())
    edges[A].append(B)
    edges[B].append(A)
for i in range(1, N + 1):
    edges[i].sort()

for i in range(1, N + 1):
    if len(edges[i]) == 0:
        print(0)
    else:
        print(len(edges[i]), *edges[i])
```

```python
N = int(input())
P = list(map(int, input().split()))
Q = []
for i in range(N - 1, 0, -1):
    a = P[i]
    b = P[i - 1]
    Q.append(a)
    if b < a:
        continue
    else:
        ans = P[:i - 1]
        target = b
        break
Q.sort()
for q in Q:
    if q < target:
        s = q
    else:
        break
Q.remove(s)
Q.append(target)
Q.sort(reverse=True)
ans = ans + [s] + Q
print(*ans)
```

```python
# C(prev_permutation関数を定義)
N = int(input())
P = list(map(int, input().split()))

def prev_permutatinon(P):
    for i in reversed(range(N - 1)):
        if P[i] < P[i + 1]:
            continue
        j = N - 1
        while P[i] < P[j]:
            j -= 1
        P[i], P[j] = P[j], P[i]
        return P[:i+1] + P[:i:-1]
    return False

print(*prev_permutatinon(P))
```

```python
from collections import Counter
N = int(input())
A = list(map(int, input().split()))

def primeFactorization(x):
    """
    xを素因数分解
    """
    factors = []
    if x == 1:
        return [1]
    i = 2
    while i ** 2 <= x:
        if x % i == 0:
            factors.append(i)
            x //= i
        else:
            i += 1
    if x != 1:
        factors.append(x)
    return factors

factors = [primeFactorization(a) for a in A]
CntF = []
for i in range(N):
    CntF.append(Counter(factors[i]))
d2 = 10 ** 5
d3 = 10 ** 5
for i in range(N):
    if 2 in CntF[i]:
        d2 = min(d2, CntF[i][2])
    else:
        d2 = 0
        break
for i in range(N):
    if 3 in CntF[i]:
        d3 = min(d3, CntF[i][3])
    else:
        d3 = 0
        break
ans = 0
for i in range(N):
    if 2 in CntF[i]:
        ans += max(d2, CntF[i][2]) - min(d2, CntF[i][2])
        CntF[i][2] = d2
    if 3 in CntF[i]:
        ans += max(d3, CntF[i][3]) - min(d3, CntF[i][3])
        CntF[i][3] = d3
        
target = []
for i in range(N):
    tmp = 0
    for c in CntF[i]:
        tmp += c * CntF[i][c]
    target.append(tmp)

if len(set(target)) == 1 or (1 in set(target) and len(set(target)) == 2):
    print(ans)
else:
    print(-1)
```

```python
N = int(input())
A = list(map(int, input().split()))

a2 = [0] * N
a3 = [0] * N
se = set()
for i in range(N):
    X = A[i]
    while X % 2 == 0:
        X //= 2
        a2[i] += 1
    while X % 3 == 0:
        X //= 3
        a3[i] += 1
    se.add(X)

if len(se) != 1:
    print(-1)
    exit()

ans = 0
m2 = min(a2)
m3 = min(a3)
for i in range(N):
    ans += (a2[i] - m2) + (a3[i] - m3)
print(ans)
```
