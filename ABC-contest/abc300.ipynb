```python
H, W = map(int, input().split())
A = [list(input()) for _ in range(H)]
B = [list(input()) for _ in range(H)]

def judge(s, t):
    for i in range(H):
        for j in range(W):
            if A[(i - s) % H][(j - t) % W] == B[i][j]: # 縦が下シフト、横が右シフトのするときの一致条件
                continue
            else:
                return False
    return True

for s in range(H):
    for t in range(W):
        if judge(s, t):
            print('Yes')
            exit(0)
print('No')
```

```python
def eratosthenes(N):
    """
    N以下の素数を全列挙
    """
    is_prime = [True] * (N + 1)
    is_prime[0] = False
    is_prime[1] = False

    i = 2
    while i * i <= N:
        if is_prime[i]:
            for j in range(2 * i, N + 1, i):
                is_prime[j] = False
        i += 1

    return [i for i in range(N + 1) if is_prime[i]]

N = int(input())
primes = eratosthenes(1000000)
M = len(primes)

ans = 0
for i in range(M):
    a = primes[i]   
    if a ** 2 >= N:
        break
    for k in range(i + 2, M):
        c = primes[k]
        if (a ** 2) * (c ** 2) >= N:
            break
        for j in range(i + 1, k):
            b = primes[j]
            if (a ** 2) * b * (c ** 2) <= N:
                ans += 1
            else:
                break
print(ans)
```
